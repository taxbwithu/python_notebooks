```python
del df['For Year']
```

```python
df=df.fillna(df.mean())
```

```python
df2015 = df[df['Period Ending'].str.contains("2015")]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

dfNorm = df.copy()
columns =dfNorm.columns.drop(['Unnamed: 0','Period Ending', 'Ticker Symbol']) 

results = scaler.fit_transform(df[columns])

dfNorm[columns] = results
```

```python
weights1 = np.ones_like(dfNorm['Gross Profit']) / len(dfNorm['Gross Profit'])
weights2 = np.ones_like(dfNorm['Earnings Per Share']) / len(dfNorm['Earnings Per Share'])

fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)
ax1.set_xlabel('Ustandaryzowany zysk brutto')
ax1.set_ylabel('Prawdopodobieństwo')
ax1.hist(dfNorm['Gross Profit'], bins =20, weights = weights1)

fig2 = plt.figure()
ax2 = fig2.add_subplot(1,1,1)
ax2.set_xlabel('Ustandaryzowany wskaźnik zysku na akcję')
ax2.set_ylabel('Prawdopodobieństwo')
ax2.hist(dfNorm['Earnings Per Share'], bins =20, weights = weights2)
```

```python
fig = plt.figure(figsize =(10, 7)) 
  
# Creating plot 
plt.boxplot([dfNorm['Gross Profit'],dfNorm['Earnings Per Share']]) 
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')

enc_df = pd.DataFrame(enc.fit_transform(dfNorm[['Ticker Symbol']]).toarray())

dfNorm = dfNorm.join(enc_df)
dfNorm
```
